Function that searches similar profiles to the users

```python
def similaritypreference():

  profilehashtable
  similaritylist = []
  usernamehashtable["similarity"] = similaritylist

  for usernames, dataset in profilehashtable.items():
    if usernames == usernamehashtable["username"]:
      continue
    else:
      for preference in usernamehashtable["preferences"]:
        for item, value in dataset.items():
          if item == preference:
            if usernamehashtable[preference] == profilehashtable[usernames][item]:
              similaritylist.append([usernames, preference]) #Same value from the user's preference
  return similaritylist

similaritylist = similaritypreference()
```

Function to extract the weights of the similar profiles only for the preference they have in common 

```python
def weightsextraction(ultimatelist, similaritylist):
  weightsultimatelist = []
  for username in ultimatelist:
    for tupleuser in similaritylist:
      if username == tupleuser[0]:
        preferenceuser = tupleuser[1]
        weightstuple = (profilehashtable[username]["weights"])
        for weight in weightstuple:
          if weight[0] == preferenceuser:
            impweight = weight[1]
            weightsultimatelist.append([username, preferenceuser, impweight])
  return weightsultimatelist

weightsultimatelist = weightsextraction(ultimatelist, similaritylist)
```

Function to sum the weights for each profile's preference with the weights of the user

```python
def sumweights(usernamehashtable, weightsultimatelist):
  for preference in usernamehashtable["weights"]:
    for user in weightsultimatelist:
      if user[1] == preference[0]:
        user[2] = user[2] + preference[1]
  return weightsultimatelist

print(sumweights(usernamehashtable, weightsultimatelist))
```

Function to check for duplicates, and if there are any sum their weight

```python
def duplicatesumweights(weightsultimatelist):
  c = Counter()
  for user, preference, weight in weightsultimatelist:
      c[user] += weight
  return list(c.items())
  
weightsultimatelist = duplicatesumweights(weightsultimatelist)
```

Quicksort algorithm to find the profile with highest similarity

```python
def quicksort(array):
  if len(array) < 2:
    return array
  else:
    left = []
    equal = []
    right = []
    pivot = random.choice(array)
    pivot = pivot[1]
    for i in array:
      if i[1] > pivot:
        left.append(i)
      elif i[1] < pivot:
        right.append(i)
      elif i[1] == pivot:
        equal.append(i)
    return quicksort(left)+equal+quicksort(right)
sortedlist = quicksort(weightsultimatelist)
```
